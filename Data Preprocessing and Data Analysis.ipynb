```python
import pandas as pd
import numpy as np
import seaborn as sns

# import modules
# import seaborn as sn
import matplotlib.pyplot as plt
```

```python
country = pd.DataFrame(data1['country'].value_counts().head())

country.reset_index(level=0, inplace=True)

sns.barplot(x = country['index'], y = country['country'], data = country)
```

```python
#date added vs release date
data1['date_added']= pd.to_datetime(data1['date_added'], infer_datetime_format=True)

sns.set(style='whitegrid')

sns.scatterplot(x=data1['date_added'],
                    y=data1['release_year'],
    
                    data=data1)
sns.set(style='whitegrid')
plt.show()
```

```python
# top_directors
# top_directors= top_directors.rename_axis('foo')
top_directors = pd.DataFrame(data1['director'].value_counts().head(10))

top_directors.reset_index(level=0, inplace=True)
top_directors
```

```python
#releases wrt year
release_number = pd.DataFrame(data1['release_year'].value_counts())
sns.lineplot(data = release_number)
sns.set(style="darkgrid")
plt.show()
```

```python
#no of movies vs no of tv shows

count_number = pd.DataFrame(data1['type'].value_counts())

count_number.reset_index(level=0, inplace=True)
sns.countplot(x =count_number['type'], 
              data = count_number, hue = 'index')
```
